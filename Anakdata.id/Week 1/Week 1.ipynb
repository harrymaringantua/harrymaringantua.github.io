# Best Selling Products by Amount and Value

```python
#top 20 products by quantity and finalprice
sns.set_style('whitegrid')
Top20Quan = df.groupby('NAMA_ITEM')['QTY'].agg('sum').sort_values(ascending=False)[0:20]
Top20Price = df.groupby('NAMA_ITEM')['REVENUE'].agg('sum').sort_values(ascending=False)[0:20]
#creating the subplot
fig,axs = plt.subplots(nrows=2, ncols=1, figsize = (12,12))
plt.subplots_adjust(hspace = 0.3)
fig.suptitle('Best Selling Products by Amount and Value', fontsize=15, x = 0.4, y = 0.98)
sns.barplot(x=Top20Quan.values, y=Top20Quan.index, ax= axs[0]).set(xlabel='Total amount of sales')
axs[0].set_title('By Amount', size=12, fontweight = 'bold')
sns.barplot(x=Top20Price.values, y=Top20Price.index, ax= axs[1]).set(xlabel='Total value of sales')
axs[1].set_title('By Value', size=12, fontweight = 'bold')
plt.show()
```

```python
#creating the pie chart
df.groupby('METODE_PEMBAYARAN')['ID_TRANSAKSI'].count().plot(kind = 'pie', autopct = '%.2f%%', figsize=(7,7)).set(ylabel='')
plt.title('Percentages of Transaction by Payment Methods', fontsize = 15)
plt.show()
```

```python
#creating the pie chart
df.groupby('METODE_PEMBAYARAN')['REVENUE'].sum().plot(kind = 'pie', autopct = '%.2f%%', figsize=(7,7)).set(ylabel='')
plt.title('Percentages of Revenue by Payment Methods', fontsize = 15)
plt.show()
```

```python
#creating the pie chart
df.groupby('Day of week')['REVENUE'].sum().plot(kind = 'pie', autopct = '%.2f%%', figsize=(7,7)).set(ylabel='')
plt.title('Percentages of Sales Value by Day of Week', fontsize = 15)
plt.show()
```

```python
#filtering customers by the total finalprice
Top10Customers = df.groupby(['CUSTOMER_ID','METODE_PEMBAYARAN'])['REVENUE'].sum().sort_values(ascending=False)[0:10]
#creating the barplot
plt.figure(figsize=(8,5))
sns.barplot(x=Top10Customers.values, y=Top10Customers.index).set(xlabel='Total Value',ylabel='CustomerID')
plt.suptitle('Top10 Customers and Payment Methods by Sales Value', fontsize = 15)
plt.show()
```

```python
#filtering customers by the total finalprice
Top10Items = df.groupby(['NAMA_ITEM','METODE_PEMBAYARAN'])['REVENUE'].sum().sort_values(ascending=False)[0:60]
#creating the barplot
plt.figure(figsize=(10,9))
sns.barplot(x=Top10Items.values, y=Top10Items.index).set(xlabel='Total Value',ylabel='Item')
plt.suptitle('Top10 Items and Payment Methods by Sales Value', fontsize = 15)
plt.show()
```

```python
#grouping customers by the number of their visits and separating them
MostRepeat = df.groupby(['CUSTOMER_ID','METODE_PEMBAYARAN'])['ID_TRANSAKSI'].nunique().sort_values(ascending=False)
rep = MostRepeat[MostRepeat != 1].values
nrep = MostRepeat[MostRepeat == 1].values
ser = pd.Series([len(rep)/(len(rep)+len(nrep)),len(nrep)/(len(rep)+len(nrep))], index=['Repeat Customers','One-time Customers'])
#creating the subplot
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize= (15,5), gridspec_kw= {'width_ratios':[3,1]})
plt.subplots_adjust(wspace=0.2)
sns.barplot(x=MostRepeat[0:10].values, y=MostRepeat[0:10].index, ax=ax1).set(xlabel='Number of Transactions(Repeats)',ylabel='CustomerID')
ser.plot(kind='pie', autopct='%.2f%%', ax=ax2).set(ylabel='')
plt.suptitle('Top Repeat Customers', fontsize=15)
plt.show()
```

```python
#grouping customers by the number of their visits and separating them
MostRepeat = df.groupby(['CUSTOMER_ID','NAMA_ITEM'])['ID_TRANSAKSI'].nunique().sort_values(ascending=False)
rep = MostRepeat[MostRepeat != 1].values
nrep = MostRepeat[MostRepeat == 1].values
ser = pd.Series([len(rep)/(len(rep)+len(nrep)),len(nrep)/(len(rep)+len(nrep))], index=['Repeat Customers','One-time Customers'])
#creating the subplot
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize= (15,5), gridspec_kw= {'width_ratios':[3,1]})
plt.subplots_adjust(wspace=0.2)
sns.barplot(x=MostRepeat[0:10].values, y=MostRepeat[0:10].index, ax=ax1).set(xlabel='Number of Transactions(Repeats)',ylabel='CustomerID')
ser.plot(kind='pie', autopct='%.2f%%', ax=ax2).set(ylabel='')
plt.suptitle('Top Repeat Customers', fontsize=15)
plt.show()
```

```python
HM_Data = df.pivot_table(index = 'METODE_PEMBAYARAN',columns = 'Day of week', values = 'REVENUE', aggfunc='sum')
plt.figure(figsize = (10,6))
sns.heatmap(HM_Data, cmap = 'vlag').set(xlabel='', ylabel='')
plt.title('Sales Value by Payment Methods and Day of Week', fontsize = 15)
plt.show()
```

```python
HN_Data = df.pivot_table(index = 'NAMA_ITEM',columns = 'Day of week', values = 'QTY', aggfunc='sum')
plt.figure(figsize = (10,6))
sns.heatmap(HN_Data, cmap = 'vlag').set(xlabel='', ylabel='')
plt.title('Sales Value by Item and Day of Week', fontsize = 15)
plt.show()
```
